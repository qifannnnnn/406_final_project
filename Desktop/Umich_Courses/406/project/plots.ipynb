```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns # sns.pairplot
# PdfPages("ssa_spectrum.pdf")
# pdf.savefig()
# pdf.close()
```

```python
dt = pd.read_csv('data/merged.csv')
```

```python
da = dt[['total_cases_per1000','death_rate','rate','low_edu','density','poverty','white_total','median_household_income_2018',
        'percent','population']]
da.corr() 
# percent; white_total; income; density; death_rate
# death_rate; density; rate
```

```python
# plot 1 poverty; income; low_edu;  employment_rate; white_total

pdf=PdfPages("cor_plot.pdf")
p1 = da[['poverty','median_household_income_2018','low_edu','white_total']]
p1 = p1.loc[p1.low_edu<=100,]
p1.columns = ['poverty rate(%)','median household income(USD)',
              'higher school education or less(%)','white population proportion(%)']
sns.pairplot(p1, corner=True,
             plot_kws=dict(marker="+", linewidth=1))
pdf.savefig()
pdf.close()
```

```python
# plot 2 percent; white_total; income; density; death_rate
p2 = da[['percent','white_total','median_household_income_2018']]
sns.pairplot(p2, corner=True,
             plot_kws=dict(marker="+", linewidth=1))
```

```python
# plot 3 cases poverty white total income
p3 = da[['total_cases_per1000','low_edu','white_total']]
sns.pairplot(p3, corner=True,
             plot_kws=dict(marker="+", linewidth=1))
```
